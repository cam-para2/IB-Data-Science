## 2b) Fit an exponential model to the data 

From our knowledge of how the virus spreads, we know that the number of infections, hospital admissions etc. should (roughly) follow an exponential curve. We would therefore like to fit a model of the form $y = c_1 e^{c_2 x}$, where $y$ is the number of admissions on day $x$. 

Note that this is a linear model on a log-scale for $y$. That is, $\log y = \log c_1 +  c_2 x$.

- Fit a linear model for $\log(\text{yvals})$ vs $\text{xvals}$, and print the values of $c_1$ and $c_2$
- Plot the fit $y = c_1 e^{c_2x}$ along with the scatterplot of the data

```python
# I've pasted in my polyreg from the previous notebook for convenience
def polyreg(data_matrix, k):
    N, _ = data_matrix.shape
    xs = data_matrix[:, 0]
    ys = data_matrix[:, 1]
    # Create the generalized matrix X.
    Xn = np.zeros((N, k+1))  # Define the size of Xn in advance for memory reasons.
    for i in range(k+1):
        Xn[:, i] = xs ** i 
    
    beta_vector = np.linalg.lstsq(Xn, ys, rcond=None)[0]

    fitted_ys_vector = Xn.dot(beta_vector)
    residuals_vector = ys - fitted_ys_vector
    return beta_vector, fitted_ys_vector, residuals_vector

log_xs = np.log(xvals)
log_ys = np.log(yvals)

log_linfit = polyreg(np.column_stack((log_xs, log_ys)), 1)
c1, c2 = log_linfit[0]
print(f"{c1=:2f} {c2=:2f}")

fitted_ys = np.exp(log_linfit[1])

plt.scatter(xvals, yvals, label="True data")
plt.plot(xvals, fitted_ys, label="Exponential fit")
plt.legend()
plt.show()
```
